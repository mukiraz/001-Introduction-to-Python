### Complex numbers

Not only real numbers, Python can also handle complex numbers and their associated functions using the "cmath" file. Complex numbers have uses in many mathematical applications. Python provides useful tools to handle these.

A complex number is represented by "x + yi". Python converts the real numbers x and y to complexes using the complex(x, y) function. The real part can be accessed using the real() function and the imaginary part can be represented with imag().

In mathematics, you cannot take the square root of -1. But

The virtual unit that provides the property is called i.

Complex numbers are frequently used in engineering. R-L-C circuits in electrical engineering can be given as an example. In a circuit where alternating current is applied, the resistance is called impedance and has a real value. When a capacitor is used in the circuit, this time the resistance is called capacitance and corresponds to the positive imaginary value.

We can use the Cartesian coordinate plane to understand complex numbers.



Complex number plotted with a point (red) and position vector (blue) in a complex plane; Also denoted as a+ib or a+bi. It is rectangular and represents a point.

```python
complex(3,4)
```

### Variables in Complex Numbers

We can also assign complex numbers to variables.

```python
c = complex(3,4)
```

If we want to get the real part of the complex number, we use the .real method.

```python
c = complex(3, 4)
```

```python
c.real **2 # square of the real part
```

If we want to get the imaginary part of the complex number, we use the .imag method.

```python
c.imag
```

If we want to find the conjugate of a complex number, we use the .conjugate method.


```python
c.conjugate()
```
