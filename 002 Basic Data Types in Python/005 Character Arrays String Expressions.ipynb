# Character Arrays (Strings)

Character arrays or strings are the same **texts** that we use in real life. In fact, string expressions are sequences of characters. For example, this sentence consists of character sequences.

### Creating a String

We create strings with "" (quote marks).

#### Creating with double quotes

```python
"Murat Ugur KIRAZ"
```

#### Creating with Single Quotes

```python
'Murat Ugur KIRAZ'
```

#### Creating with triple quotes

```python
"""Murat Ugur KIRAZ"""
```

The character strings we create with triple quotes do not have to be on the same line. Allows multiple lines.

```python
"""
We can write
string expression
in multiple lines.
"""
```

Note: If we started with double quotes, we must close it with double quotes, if we started with single quotes, we must close it with single quotes and define the string.

```python
"Murat' # This code gives an error
```

```python
"Murat" # This code does not give an error
```

```python
"Dwayne "The Rock" Johnson can’t escape his origins as a wrestler." #This code gives an error
```

### How are string expressions defined on the computer?

String expressions are kept in RAM as a series of characters. Each character is represented as the code of an ASCII or Unicode character. ASCII is an 8-bit coding system. Unicode is a 16-bit or 32-bit coding system.

As an example:

```python
s = "Hello"
```

When we define a string as , the variable s is a string object. In computer language, an object is a data structure that occupies space in RAM and contains methods (functions) and data. In a string variable, each character has a mathematical value as data.

s = In the string "Hello", each letter corresponds to the number specified in the ASCII or Unicode code table and is kept in RAM.



Python is a C-based language. String expressions in C language are defined as character arrays as shown in the figure. In the C language, there is a null value, that is, 00h, at the end of every string expression.

The s variable actually stores the address of the first character of the string expression in RAM. Other characters are listed side by side. When we call the variable s, the computer starts reading from the first address of the string expression and continues reading until it sees the null expression. In other words, the computer detects the end of a string expression with an 8-bit zero.

Therefore, we cannot change any character of string expressions. Even if we want to change it, we need to copy the relevant string to another area in RAM. String operations are costly.

Here, the variable s is actually a string object. It contains methods.

If we press the TAB key after saying s., we can see the methods of the s object.




```python
s[2] = "f" # This code throws an error. With this code, we wanted to make the 3rd character from the left of the variable s "f".
```
