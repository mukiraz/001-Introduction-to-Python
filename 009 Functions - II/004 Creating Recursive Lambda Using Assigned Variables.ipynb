# Creating Recursive Lambda Using Assigned Variables

One method of creating recursive lambda functions involves assigning the function to a variable and then referencing that variable within the function. A common example of this is the recursive calculation of the factorial of a number, as shown in the following code:

```python
lambda_func = lambda i : 1 if i == 0 else i * lambda_func(i-1)
print(lambda_func(4)) # 4 * 3 * 2 * 1 = 12 * 2 = 24
```

The lambda function is passed a value (4) via variable assignment, which it evaluates to and returns 1 if 0, or otherwise, another calculation by the lambda function of the value -1 (i-1) returns the current value i*.
