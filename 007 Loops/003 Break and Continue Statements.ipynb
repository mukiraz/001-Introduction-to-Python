## break statement

When the break statement is executed within a loop, the control flow is immediately broken from the loop.

The break statement **only** terminates the loop it is in. If there are nested loops and break is used in the innermost loop, only the innermost loop ends.

The **break** statement inside a function cannot be used to terminate loops that call that function.

### An example of a while loop

```python
i = 0
while i < 7:
    print(i)
    if i == 4:
        print("I quit the loop")
        break
    i += 1
```

Above, the number i took the values ​​of 0, 1, 2, 3, 4 respectively. When the expression 4==4 was encountered, it switched to the block of the if statement. After writing "I am leaving the loop" on the screen, the loop was terminated by executing the break statement. The break statement only appears inside loops.

### continue statement
The *continue* expression is a slightly less used expression than *break*. Its meaning is as follows;
            
        When the loop encounters the continue statement anywhere and at any time, it executes the remaining operations.
        Without doing so, he returns directly to his block.

## Flowchart of Continue Statement



```python
liste = [1,2,3,4,5,6,7,8,9] # Let's use continue


for i in liste:
    if (i == 3 or i == 5):
        print("   pass the value is 3 or 5")
        continue
    print("i:",i)
```

### Usage of Continue Statement

Loops in Python automate and repeat tasks efficiently. But sometimes, there may arise a condition where you want to exit the loop completely, skip an iteration or ignore that condition. These can be done by loop control statements. Continue is a type of loop control statement that can alter the flow of the loop.
