## 6. Egzersiz

1937'de Lothar Collatz adlı bir Alman matematikçi, şu şekilde tanımlanabilecek ilgi çekici bir hipotez formüle etti (hala kanıtlanmadı):

1. Negatif olmayan ve sıfır olmayan herhangi bir tam sayıyı alın ve onu c0 olarak adlandırın;
2. Eğer çift ise yeni c0'ı c0 ÷ 2 olarak değerlendirin;
3. Aksi takdirde, eğer tek ise, yeni bir c0'ı 3 × c0 + 1 olarak değerlendirin;
4. c0 ≠ 1 ise 2. noktaya geçin.

Hipotez, c0'ın başlangıç değeri ne olursa olsun her zaman 1'e gideceğini söylüyor.

Elbette herhangi bir doğal sayıya ilişkin hipotezi kanıtlamak için bilgisayar kullanmak son derece karmaşık bir iştir (hatta yapay zeka gerektirebilir), ancak bazı bireysel sayıları kontrol etmek için Python'u kullanabilirsiniz. Belki hipotezi çürütecek olanı bile bulacaksınız.
