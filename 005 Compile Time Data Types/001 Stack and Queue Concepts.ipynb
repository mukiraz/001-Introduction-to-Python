# Stack concept

The stack is a useful data structure in programming. It's like a stack of CDs stacked on top of each other.

If we want to add a new CD to a CD stack like above, we have to add it to the top. If we want to buy a CD from this stack of CDs, we will buy the one at the top.
If we want to get to the bottom CD, we have to remove all the CDs including the bottom one, take the bottom one and then put the rest back in. We call such a structure **Last in, First Out**, that is, LIFO (Last in first out).

Data sets that work with LIFO logic are called **stack**, that is, **stack** in programming.

In the image above, the 3rd data was added last. However, it was removed first. Therefore, it follows the Last In First Out (LIFO) principle.

In programming, putting an item at the top of the stack is called **push**, and removing an item is called **pop**.
