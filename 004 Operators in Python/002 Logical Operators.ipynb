## *and (and)* logic gate coding example

```python
a = False
b = False

print(a and b)

a = True
b = False

print(a and b)

a = False
b = True

print(a and b)

a = True
b = True

print(a and b)
```

## *or (or)* logic gate coding example

```python
a = False
b = False

print(a or b)

a = True
b = False

print(a or b)

a = False
b = True

print(a or b)

a = True
b = True

print(a or b)
```

## *not (not)* logic gate coding example

```python
a=False

print(a)
print(not a)

```

## Comparison Operators

There will often be decision stages in the Python codes we write. We will encounter situations such as one value being equal to another value, being less than, greater than, or not equal to another value. We will use *Comparison Operators* to perform these operations.

During coding, when we need to compare a variable with a constant or another variable; it would be correct to use it as:

| Operator| Description| Usage| Result| 
| --- |  --- |  --- | --- | 
| == | Is equal? | 5 == 5 | True |
| != | Is not equal? | 5 != 5 | False |
| > | Is greater? | 5 > 5 | False |
| < | Is smaller? | 5 < 5 | False |
| >= | Is bigger or equal? | 5 >= 5 | True |
| <= | Is samller or equal? | 5 <= 5 | True |



The above expressions are represented in code as follows.

```python
a = 5
b = 5

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

### Comparing the results of comparison operators with logical operators

We can compare the results of comparison operators with logical operators.

For example;
**Is the number 5 greater than both the number 4 and the number 6?**
It is expected that both conditions will be met. However, since the number 5 is less than 6, it does not meet the second condition. So the result of the expression will be *FALSE*.

Or

**Is the number 5 greater than the number 4 or the number 6?**
In this example, it is sufficient for only one of the two conditions to be true. Therefore, the result of the expression will be *TRUE*.

Let's write the Python codes for the two examples above;

```python
a=5
b=4
c=6

# Is the number 5 greater than both the number 4 and the number 6?
print((a>b) and (a>c))

# Is the number 5 greater than the number 4 or the number 6?

print((a>b) or (a>c))
```
