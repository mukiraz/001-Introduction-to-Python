## Why Will We Use Jupyter in Our Courses?

Jupyter Notebook is an open source program that provides an interactive environment for data science and scientific computing. Supports Python, R, Julia, and many other programming languages.

Jupyter Notebook works like a notebook. Each notebook page consists of one or more cells. Cells can contain code, text, Markdown, and images. When code cells are run, the results are printed at the bottom of the cell.

Jupyter Notebook is a popular tool for data science and scientific computing because:

* **Interactive:** Jupyter Notebook allows you to run code cells one by one and see the results immediately. This is ideal for research and development.
* **Its content is rich:** Jupyter Notebook, you can combine different content types such as code, text, Markdown, and images in the same notebook. This makes it easy to create documentation and descriptions for your data science projects.
* **Shareable:** You can easily share your Jupyter Notebooks with others. This is ideal for collaboration and repeatability.

Jupyter Notebook is widely used by data scientists, data analysts, researchers, and other professionals.

## Running Jupyter Notebook

After installing Anaconda Python Distribution, find and run Jupyter Notebook from windows programs.

Jupyter is an application that runs on the web browser. Immediately after running Jupyter, you will see a command screen. This command screen is as follows.

If you selected the option below when installing the Anaconda software, this will show the user folder on your computer. Jupyter opens directly here.



After the command line runs, your default web browser opens. If it does not open, enter the address line of your web browser in the red section in the picture above.
- http://localhost:8888/?token=token
or 
- http://127.0.0.1:8888?token=token

You can copy and paste the address. Jupyter creates a new token every time it starts. This token is for security. Without this token, the Jupyter application will not be visible in the web browser.

The Jupyter application will start as follows.

7. Jupyter terminates the program.
